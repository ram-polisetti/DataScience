```python
import pandas as pd
```

```python
meta = pd.read_csv('Maang Stock Prices/META_daily.csv')
apple = pd.read_csv('Maang Stock Prices/APPLE_daily.csv')
amazon = pd.read_csv('Maang Stock Prices/AMAZON_daily.csv')
netflix = pd.read_csv('Maang Stock Prices/NETFLIX_daily.csv')
google = pd.read_csv('Maang Stock Prices/GOOGLE_daily.csv')
```

```python
dfs = [apple, meta, google, amazon, netflix]
```

```python
for df in dfs:
    df['PreviousDayClosingPrice'] = df['Close'].shift(1)
```

```python
meta.head()
```

```python
for df in dfs:
    df['ClosingPriceChange'] = df['Close'] - df['PreviousDayClosingPrice']
```

```python
meta.head()
```

```python
for df in dfs:
    df['ClosePricePercentChange'] = df.Close.pct_change()
```

```python
for df in dfs:
    df['MA50'] = df.Close.rolling(50).mean()
    df['MA200'] = df.Close.rolling(200).mean()
```

```python
meta.head()
```

```python
for df in dfs:
    df['PreviousDayVolume'] = df.Volume.shift(1)
```

```python
for df in dfs:
    df['VolumeChange'] = df['Volume'] - df['PreviousDayVolume']
```

```python
for df in dfs:
    df['VolumePercentChange'] = df.Volume.pct_change()
```

```python
apple.to_csv('apple.csv')
meta.to_csv('meta.csv')
google.to_csv('google.csv')
amazon.to_csv('amazon.csv')
netflix.to_csv('netflix.csv')
```
